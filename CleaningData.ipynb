```python
import pandas as pd
```

```python
df = pd.read_csv("gradedata.csv")
print(df.to_string())
```

```python
#EMPTY CELLS
#Remove Rows
df.dropna(inplace = True) #Remove all rows with NULL values
print(df.to_string())
```

```python
#EMPTY CELLS
#Remove Empty Value
df.fillna(130,inplace = True) #Replace NULL values with the number 130
```

```python
#EMPTY CELLS
#Remove only For Specified Collumns
df.fillna({'age': 25}, inplace = True) #Replace NULL values in the "age" columns with the number 25
```

```python
#EMPTY CELLS
#Replace using Mean, Medain or Mode
x = df["age"].mean()
y = df["age"].median()
z = df["age"].mode()[0]
df.fillna({'age': x}, inplace = True) #Calculate the MEAN, and replace any empty values with it
print(df.to_string())
```

```python
#WRONG FORMAT
df['Date'] = pd.to_datetime(df['Date'], format = 'mixed') #data ko co date nen ko hien, sua du lieu ve dung format
df.dropna(subset = ['Date'], inplace = True)
print(df.to_string())
```

```python
#WRONG DATA
df.loc[7, 'age'] = 20

for x in df.index:
    if df.loc[x,"age"] > 18:
        df.loc[x,"age"] = 18 #Thay the

for x in df.index:
    if df.loc[x,"age"] > 17:
        df.drop(x, inplace = True) #Xoa du lieu
print(df.to_string())
```
