```python
import pandas as pd
```

```python
#Series
a = [4,6,3]
myvar = pd.Series(a)
myvar1 = pd.Series(a, index = ["x","y","z"])
print(myvar)
print(myvar1["y"])
print(myvar[0])
```

```python
calories = {'day1': 420, 'day2': 400, 'day3': 480}
myvar = pd.Series(calories)
myvar1 = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
print(myvar1)
```

```python
#DataFrame
mydataset = {
    'cars': ["Ford","Honda","BMW"],
    'colors': ["red","green","black"],
    'passing': [5,7,4]
}
myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
data = {
    'Name':["Duy","Minh","Tam","Diem"],
    'Gender':["Male","Male","Female","Female"],
    "Age":[21,22,20,19]
}
data1 = pd.DataFrame(data)
data2 = pd.DataFrame(data, index = ["No1","No2","No3","No4"])
#print(data1)
print(data1.loc[0])
print(data1.loc[[0,1]]) #using [], the results is pandas DataFrame
print(data2.loc["No4"])
```
